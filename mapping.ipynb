**[Hint]**<br>
* city 변수는 시도코드와 시도명이 담겨있는 string 변수입니다.
* 시도코드와 시도명은 Space로 구분이 되어있습니다.
* 각각의 [시도코드, 시도명]은 Enter로 구분이 되어있습니다.

**※ city 변수는 수정 불가합니다.**
